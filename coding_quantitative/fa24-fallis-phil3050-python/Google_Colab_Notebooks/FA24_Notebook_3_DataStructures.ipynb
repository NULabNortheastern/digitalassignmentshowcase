# Python Data Structures

So far, we've seen a few of Python's inbuilt data types: integers, floats, and strings. This notebook will discuss data structures. These help us store many values inside of a single variable. We will mainly focus on [tuples](https://constellate.org/docs/key-terms/#tuple) and [dictionaries](https://constellate.org/docs/key-terms/#dictionary) as you will use these in your Dunkin assignments. At the end, we will also briefly touch on [lists](https://constellate.org/docs/key-terms/#list) and [sets](https://www.w3schools.com/python/python_sets.asp).  

The fundamental difference between a tuple and a dictionary is that a tuple stores items in sequential order (starting from 0) while a dictionary stores items in key/value pairs. When we want to retrieve one or more items from a tuple, we use an index number or a set of index numbers as a reference. When we want to retrieve an item from a dictionary, we supply a key that returns the value associated with that key. Another important difference is that tuples are **immutable**, which means they can't be changed, but dictionaries are **mutable**, which means they can be changed after they are initialized.

Each of these data structures can be beneficial depending on what kind of data we are working with, and what we intend to do with the data.

For your class assignments using Python, you will be asked to work with data that is stored in dictionaries and tuples. We will purposefully be spending more time on dictionaries and tuples to get you ready for these assignments.

## Tuples

A [tuple](https://www.geeksforgeeks.org/tuples-in-python/) can store anywhere from zero to millions of items. The items that can be stored in a tuple include the data types we have already learned: integers, floats, and strings—and a tuple may contain different data types. A tuple assignment statement takes the form:

`my_tuple = (item0, item1, item2, item3)`

The items in a tuple are separated by commas and enclosed in parentheses.

```python
# Code cell 1
# This is a tuple containing integers.
my_favorite_numbers = (17, 19, 100)
print(my_favorite_numbers)
```

```python
# Code cell 2
# This is a tuple containing strings.
my_favorite_philosophers = ('Margaret Cavendish', 'Mary Wollstonecraft', 'Hannah Arendt')
print(my_favorite_philosophers)
```

Both `my_favorite_numbers` and `my_favorite_philosophers` have three items, but we could have also initialized them with no items by leaving the parentheses empty:

`my_favorite_numbers = ()`

or we could have included many more items. Each item has an [index number](https://constellate.org/docs/key-terms/#index-number) that represents their position in the tuple. The first item is at the index 0, the second item is at 1, the third item is at 2, etc. In the `my_favorite_philosophers` tuple, `'Hannah Arendt'` is at the index number 2.

To retrieve an item from a tuple, we put the name of the tuple, followed by the index number for the item we want to retrieve in square brackets:

`tuple name[index number]`

```python
# Code cell 3
# Retrieve an item in a tuple.
my_favorite_philosophers[2]
```

What do you think will happen if we change the index number to 1? What about 3?

We can retrieve a group of consecutive items from a tuple using a [slice](https://constellate.org/docs/key-terms/#slice). Think of a slice like cutting a slice of cake where the cake is the tuple and the slice is some consecutive set of items we want from the tuple. We create a **slice** by indicating a starting and ending index number, separated by a colon:

`tuple name[starting index number: ending index number]`

The slice contains all the items between our starting and stopping index number.

```python
# Code cell 4
# Take a slice of a tuple.
historical_periods = ('Classical Antiquity',
                      'Early Middle Ages',
                      'High Middle Ages',
                      'Late Middle Ages',
                      'Early Modern Period',
                      'Late Modern Period',
                      'Contemporary History')
historical_periods[3:5]
```

Notice that the second index in a slice is the stopping point. A way to check this is by subtracting the indexes in your head (5 - 3 = 2 items). While taking a slice of a tuple may not be nessecary for your assignments, it can be especially useful when you want to retrieve a subset of the data stored in a tuple.

It is not uncommon for tuples to be hundreds or thousands of items long. If you want to know the length of a tuple, you can use the `len()` function.

```python
# Code cell 5
# Use the len() function to discover the number of items in the tuple.
len(historical_periods)
```

## Dictionaries

Like a tuple, a [dictionary](https://constellate.org/docs/key-terms/#dictionary) can hold many values within a single variable. We have seen that the items of a tuple are stored in a strictly-ordered fashion, starting from the index number 0. In a dictionary, each [value](https://constellate.org/docs/key-terms/#key-value-pair) is stored in relation to a descriptive identification [key](https://constellate.org/docs/key-terms/#key-value-pair) forming a [key/value pair](https://constellate.org/docs/key-terms/#key-value-pair). This structure makes dictionaries very useful, because you can supply a key and receive a value without needing to refer to a specific index number. You are not allowed to have duplicate keys in Python dictionaries since keys are used for identification purposes; each key can be used only once.

Whereas a tuple is typed with parentheses `()`, a dictionary is typed with curly braces `{}`.  The following examples show how key/value pairs can be used to store different kinds of data in a dictionary.

`example_dictionary = {keyA : valueA, keyB : valueB, keyC : valueC}`

`menu_dictionary = {itemA : priceA, itemB : priceB, ...}`

`gradebook_dictionary = {nameA : gradeA, nameB : gradeB, ...}`

Another important difference between dictionaries and tuples is that dictionaries are **mutable**, meaning that data in dictionaries can be changed after the dictionary is initialized. Tuples, on the other hand, cannot be changed after being initialized.

The values in dictionaries can be of any data type and you can have a mix of different data types within a single dictionary. The keys can composed of most data types, including integers, floats, strings, and tuples. For example, strings could be used as alphabetic names for your keys or integers could be used as ID numbers for your keys, etc.

###Writing a generalized function to modify a data container

You may want to use a function multiple times to change a data container like a dictionary. You can write a generalized function, a function that is written to accept different inputs.

Recall that writing a function first requires us to write the header, where we will name the function and specify any parameters. Then we will write the body, where we will put our statements that define what the function will do. Lastly, we will write a `return` statement if we want the function to return a specific output.

You may want a generalized function to add new key/value pairs to an existing dictionary. For example, let's add a new menu item to the `breakfast_menu`. Let's look at a specific example of how to add Pancakes and their price to the menu, and then write our generalized function.

### `for` Loops and Dictionaries
In last week's workshop, we learned how to use for loops to iterate over a list. You can also write `for` loops to iterate over tuples and dictionaries.

By default, if you put a dictionary into a `for` loop, it will iterate over the keys in that dictionary. Here's an example of how that works.

```python
# Code cell 25
short_menu ={
 'Breakfast Sandwich': 9.75,
 'Croissant Breakfast Sandwich': 11.00,
 'Biscuit Sandwich': 9.00}

for key in short_menu:
    print(key)
```

##Lists

While the assignments do not focus on lists, it can be helpful to understand them. In our last workshop, we used iterations to execute commands on each element within a list. We saw that lists can store integers, float values, and strings. We also saw that lists are changeable, or **mutable**.

Lists can store anywhere from zero to millions of items, and they store these items in order.

To initialize a list, we use a list definition statement, in the following format:

`my_list = [item0, item1, item2, item3, ...]`

Like data in tuples, data in lists are stored in a set order and individual items can be referenced by the item's [index number](https://constellate.org/docs/key-terms/#index-number), or position in the list.

```python
# Code cell 29
# Let's initialize a list.
my_list = [85, 89, 81, 90, 96, 100]
```

So, if we want to retrieve only one item in `my_list`, we would use the following code block:

`my_list[index number]`

```python
# Code cell 30
# Retrieve an item in a list.
my_list[1]  #We want to retrieve the second item in the list. Since the index number begins at 0, we refer to index number 1 to retrieve the second item in the list.
```

What happens when you change the index number to 3? Try it for yourself!

Like retrieving slices of tuples, we can also retrieve slices of lists using the index numbers. To do this, we use the following code:

`my_list[starting index number : ending index number]`

```python
# Code cell 31
#Retrieve a slice from a list.
my_list[1:4]
```

Like when we retrieve a slice from a tuple, the second index number is the stopping point.

We saw last week that we can modify lists. To replace an individual item in a list, we can refer to the item's index number and use an assignment statement to re-assign the item's value. Let's take a look at this.

For example, let's say my_list represents your grades over the past few assignments. You did some extra credit to get 5 points added to your lowest grade. Let's use the index number to specify the lowest value in `my_list` and tell Python to re-assign it with a new value.

```python
# Code cell 32
my_list = [85, 89, 81, 90, 96, 100]  # The lowest value is at index number 2.
my_list[2]= (81+5)  # Re-assign the value to be 81+5.
my_list  # Recall my_list to double-check the change was executed.
```

We can also use the `len()` function on lists to check how long our lists are. This is useful if you have been making changes to your lists and want to keep track of how many items are stored in your list.

```python
# Code cell 33
# Check how many items are in your list with the len() function.
len(my_list)
```

We can also use a `for` loop to change the numbers in our list by replacing each number at each index with a different number.

```python
# Code cell 34
index_number = 0  # Set the index number to 0 so the for loop starts with the first value in our list.
for number in my_list:
  print("index_number:", index_number)  # Print the index number for that iteration of the for loop.
  print("original number:", my_list[index_number])  # Print the original number at that index.
  multiplied_number = my_list[index_number] * 2  # Multiply the original number by two.
  my_list[index_number] = multiplied_number  # Replace the old number with the multiplied number in the list.
  print("multiplied number:", my_list[index_number])
  index_number = index_number + 1  # Increase the index number by 1 so that the next iteration changes the next item in the list

print("Multiplied list:", my_list)
```

## Sets

Our focus here has been on tuples and dictionaries, but we are briefly going to discuss one more data type: **sets**. Sets are not covered in the assignments but can still be useful to understand. Sets are the last of the four built-in data types in Python that can store collections of data. To recap: these are tuples, dictionaries, lists, and sets.

Sets are expressed like so:

`example_set = {"item1", "item2", "item3"}`

Sets are **unordered** and **unindexed**. That means that the items in a set could be in a different order any time you reference them. You cannot refer to the items in a set by index numbers, as you do with tuples. You'll also note from our example that the items in a set are not keyed, as with dictionaries. Because sets are unindexed, you are not allowed to have duplicate values.

You cannot change the items in a set once the set has been initialized. However, you can remove items or add new ones. Taking our generic example, this means that we could remove `item3` or add `item4`, but we couldn't change `item1` to `item_one`.

Sets may contain a mix of different data types. However, sets cannot contain **mutable** data types like dictionaries or lists.

```python
# Code cell 35
# Initialize an example set.
my_favorites = {"green", 19, "lily of the valley", "blackberries", "cookie dough"}
```

```python
# Code cell 36
# We can use the `len()` function to determine the length of a set.
len(my_favorites)
```

```python
# Code cell 37
# We can also use the `type()` function.
type(my_favorites)
```

One thing that makes sets useful is that they do not allow repeat items: this means that putting items into a set is an effective way of removing any duplicates. Sets are also useful for some math operations, but those are out of the scope of this lesson.

```python
# Code cell 38
# If you do put duplicate values in a set, they will be ignored.
my_absolute_favorites = {"green", 19, "lily of the valley", "blackberries", "cookie dough","green"}
print(my_absolute_favorites)
```
